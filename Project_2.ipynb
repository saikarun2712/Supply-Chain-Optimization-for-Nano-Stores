```python
import pandas as pd

poverty.rename(columns={
    'Folio de la AGEB': 'ageb'
   }, inplace=True)
# Perform a left join
nano_poverty_merged = pd.merge(nano, poverty, on=['cve_ent', 'cve_mun', 'cve_loc','ageb'], how='left')

# Display the first few rows to check the results
nano_poverty_merged
```

```python
#dropping columns

nanopov_merged_df = nano_poverty_merged.drop(['Entidad\nfederativa', 'Nombre del municipio', 'Nombre de la localidad','Grado de Rezago Social'], axis=1)

#renaming columns

nanopov_merged_df.rename(columns={
    'Poblacin total': 'Total Population',
    'Viviendas particulares habitadas con caracteristicas': 'Total Dwellings',
   }, inplace=True)
```
