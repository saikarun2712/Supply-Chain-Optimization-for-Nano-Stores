###Retail-1:

1. `id` - **ID**: A numeric identifier for the National Statistical Directory of Economic Units (DENUE), a unique numeric key for each record in the database.
2. `nom_estab` - **Establishment Name**: The commercial name or external name that identifies the economic unit. Usually, it's easily recognized because it is visible and written on signs, facades, or lighted announcements. If there is no commercial name, then the generic name of the activity or the description of the economic activity is used.
3. `raz_social` - **Legal Name**: The legal constitution and registration of the moral person or the denomination of a legally constituted society. When the economic unit belongs to an individual, the data is not included to protect confidentiality.
4. `codigo_act` - **Activity Code**: The classification of activities developed by the economic units according to the North American Industry Classification System (NAICS) and its updates.
5. `nombre_act` - **Activity Name**: The name of the activity code according to NAICS 2018.
6. `per_ocu` - **Occupancy Period**: Refers to personnel directly hired by the business and external personnel supplied by another entity, who worked for the economic unit, subject to its direction and control, and covered at least a third of the working day.
7. `tipo_vial` - **Road Type**: The type of road where the economic unit is located (street, avenue, pathway, etc.).
8. `nom_vial` - **Road Name**: The proper noun that identifies the road.
9. `tipo_v_e_1` - **First Adjacent Road Type**: Similar to `tipo_vial`, but for the first adjacent road or crossing.
10. `nom_v_e_1` - **First Adjacent Road Name**: The name of the first adjacent road.
11. `tipo_v_e_2` - **Second Adjacent Road Type**: Road type for the second adjacent road.
12. `nom_v_e_2` - **Second Adjacent Road Name**: The name of the second adjacent road.
13. `tipo_v_e_3` - **Third Adjacent Road Type**: Road type for the third adjacent road.
14. `nom_v_e_3` - **Third Adjacent Road Name**: The name of the third adjacent road.
15. `numero_ext` - **External Number**: The numerical values that identify one or several properties on a road.
16. `letra_ext` - **External Letter**: Alphanumeric characters that identify one or several properties on a road.
17. `edificio` - **Building**: The proper noun that identifies the building where the economic unit is located.
18. `edificio_e` - **Building Level**: Identifies the floor or level of the building where the economic unit is located.
19. `numero_int` - **Internal Number**: The numerical values that identify the address of the economic unit inside a building.
20. `letra_int` - **Internal Letter**: Alphanumeric characters that identify an address inside a building.
21. `tipo_asent` - **Settlement Type**: The classification of the human settlement according to its function (neighborhood, housing development, housing unit, etc.).
22. `nomb_asent` - **Settlement Name**: The proper noun that identifies the human settlement where the economic unit is located.
23. `tipoCenCom` - **Commercial Center Type**: Type of plaza, shopping center, etc., where the economic unit is located.
24. `nom_CenCom` - **Commercial Center Name**: Name of the plaza, shopping center, etc., where the economic unit is located.
25. `num_local` - **Local Number**: The alphanumeric data corresponding to the local where the economic unit was located.
26. `cod_postal` - **Postal Code**: A numerical code consisting of five digits established by the Mexican Postal Service.
27. `cve_ent` - **State Key**: Key that identifies the statistical federal entity.
28. `entidad` - **Entity**: The proper noun that identifies the federal entity.
29. `cve_mun` - **Municipality Key**: Key that identifies the statistical municipality, and in the case of Mexico City, the keys that identify the Boroughs.
30. `municipio` - **Municipality**: The proper noun that identifies the municipality, and in the case of Mexico City, the Boroughs.
31. `cve_loc` - **Locality Key**: Key that identifies the statistical locality.
32. `localidad` - **Locality**: The proper noun that identifies the locality.
33. `ageb` - **Basic Geostatistical Area**: A territorial extension that corresponds to the subdivision of the Basic Geostatistical Areas (AGEM).
34. `manzana`

 - **Block**: The geostatistical key of the block, composed of three digits, identifying a territorial extension consisting of a group of houses, buildings, land, lots, or parcels used for residential, commercial, industrial, or service purposes.
35. `telefono` - **Telephone**: The telephone number, a sequence of digits that allows telephone communication with the economic unit.
36. `correoelec` - **Email**: The series of characters used to access an email service that enables the economic units to send and receive messages via electronic communication systems.
37. `www` - **Website**: Data that allows access to information about the economic unit on the Internet.
38. `tipoUniEco` - **Economic Unit Type**: Categories to identify if the establishment is fixed or semi-fixed.
39. `latitud` - **Latitude**: The distance between the economic unit and the equator, measured over the meridian that passes through that point.
40. `longitud` - **Longitude**: The distance between the economic unit and the Greenwich meridian, measured over the parallel that passes through that point.
41. `fecha_alta` - **Registration Date**: The date the economic unit was added to the National Statistical Directory of Economic Units and the Statistical Business Register of Mexico (RENEM).
42. `clee` - **Statistical Identification Key**: The unique statistical identification key assigned exclusively by INEGI to each of the establishments and businesses registered in RENEM, to uniquely identify them and link establishments from the same company.

```python
# Initialize the translator
translator = Translator()
# Function to translate column names
def translate_columns(columns):
    translated_columns = {}
    for column in columns:
        # Translate the column name and use the English name or the original if error occurs
        try:
            translated = translator.translate(column, src='es', dest='en').text
            translated_columns[column] = translated
        except Exception as e:
            print(f"Error translating {column}: {e}")
            translated_columns[column] = column
    return translated_columns
```

```python
# Translate the DataFrame column names
translated_columns = translate_columns(retail_1.columns)

# Rename the columns using the translated names
retail_1.rename(columns=translated_columns, inplace=True)
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Reading the CSV
df_mex = pd.read_csv('/content/drive/MyDrive/mexico_nano.csv')

# Creating a GeoDataFrame with selected columns
# Let's say you want to keep 'Column1', 'Column2', and 'Column3' alongside the coordinates
selected_columns = df_mex[['id', 'longitud', 'latitud','R_U']]
gdf = gpd.GeoDataFrame(
    selected_columns,
    geometry=[Point(xy) for xy in zip(df_mex['longitud'], df_mex['latitud'])]
)

# Setting the CRS
gdf.crs = "EPSG:4326"

# Writing to Shapefile
gdf.to_file(driver='ESRI Shapefile', filename='mexico_nano_RU.shp')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Define colors based on the 'R_U' column
color = mex_shp_RU['R_U'].map({'R': 'red', 'U': 'blue'})

# Plotting
mex_shp_RU.plot(figsize=(10, 10), color=color)
plt.show()
```

```python

# To ensure CRS is set for distance calculations (if needed, set to appropriate CRS)
mex_shp_RU = mex_shp_RU.set_crs("EPSG:4326")  # Adjust the EPSG code as necessary

# Calculate pairwise distances (This can be very computationally expensive for large datasets)
# distances = mex_shp_RU.geometry.apply(lambda g: mex_shp_RU.geometry.distance(g))
# average_distance = distances.stack().mean()

# Calculate densities
# Filter by Rural and Urban
rural_stores = mex_shp_RU[mex_shp_RU['R_U'] == 'R']
urban_stores = mex_shp_RU[mex_shp_RU['R_U'] == 'U']

# Assuming we define a 'total area' or use the total bounding box area for simplicity
total_area = mex_shp_RU.total_bounds  # This gives the bounding box: (minx, miny, maxx, maxy)
total_area_size = (total_area[2] - total_area[0]) * (total_area[3] - total_area[1])  # Area calculation might need projection correction

# Density calculation as number of stores per some unit area
density_rural = rural_stores.shape[0] / total_area_size
density_urban = urban_stores.shape[0] / total_area_size

# Output results
# print(f"Average Distance Between Stores: {average_distance} units")
print(f"Density of Rural Stores: {density_rural} stores per unit area")
print(f"Density of Urban Stores: {density_urban} stores per unit area")
```
